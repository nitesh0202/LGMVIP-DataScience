```python
#spliting data into trainset and testset
from sklearn.model_selection import train_test_split
x=data.iloc[:,:3]
y=data["Species"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=8)
```

```python
#Visualize the decision tree
from sklearn import tree
decision_tree=tree.DecisionTreeClassifier()
decision_tree.fit(x_train,y_train)
predictedTree=decision_tree.predict(x_test)

figure=plt.figure(figsize=(12,15))
_ = tree.plot_tree(decision_tree, 
                   feature_names=['sepal_length', 'sepal_width', 'petal_length','petal_width'],  
                   class_names=['Setosa', 'Versicolor', 'Virginica'],
                   filled=True)
```

```python
from sklearn import metrics
print("Accuracy of decision tree :",metrics.accuracy_score(y_test, predictedTree) )
```
